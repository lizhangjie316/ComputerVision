```python
import numpy as np
from matplotlib import pyplot as plt

def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-3,3,100)
y = sigmoid(x)
plt.figure()
plt.plot(x,y)
#plt.xlim(-3,3)

ax = plt.gca()
#ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')




plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 12
X = np.arange(12)
Y1=(1-X/float(n))*np.random.uniform(0.5,1.0,n)
Y2=(1-X/float(n))*np.random.uniform(0.5,1.0,n)

plt.bar(X,+Y1,facecolor="#9999ff",edgecolor='white')
plt.bar(X,-Y2,facecolor="#ff9999",edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x,y+0.03,'%.2f' % y,ha='center',va='bottom')  #存在顺序
    
for x,y in zip(X,Y2):
    plt.text(x,-y-0.03,'%.2f' % -y,ha='center',va='top')    


plt.xlim(-.5,n)
plt.ylim(-1.25,1.25)
#plt.xticks(())
#plt.yticks(())

plt.show()

```

# 10-3D数据展示

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax =Axes3D(fig)
#X,Y value
X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2+Y**2)
#height value
Z = np.sin(R)
#绘制曲面             #row stride 行跨度，column stride:列跨度
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=plt.get_cmap('rainbow'))
#填充等高面    z: 按z轴方向进行  offset:-2代表在z=-2那个平面进行
ax.contourf(X,Y,Z,zdir='z',offset=-2,cmap='rainbow')
#ax.contourf(X,Y,Z,zdir='x',offset=-4,cmap='rainbow')
ax.set_zlim(-2,2)

plt.show()
```

# 11-subplot 多合一显示

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure()
#将一个图像分为2*2 的四个子图，其中在第一张图上绘制：
plt.subplot(2,2,1)
plt.plot([0,1],[0,1])  #x:0->1    y:0->1

#绘制第二张小图
plt.subplot(2,2,2)
plt.plot([0,1],[0,2])

plt.subplot(2,2,3)
plt.plot([0,1],[0,3])

plt.subplot(2,2,4)
plt.plot([0,1],[0,4])

plt.show()
```

# 12-图像的分格显示 subplot in grid （这种方式会比上一个好一点）

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# method 1:subplot2grid
################################
plt.figure()   #划分为3行3列；从figure的第0行0列位置开始(即左上角)，进行划，列跨度为3  行跨度为1，
#作第一张图
ax1 = plt.subplot2grid((3,3),(0,0),colspan=3,rowspan=1)  
ax1.plot([1,2],[1,2])
ax1.set_title('ax1_title')


#作第二张图      从第1行0列的位置开始
ax2 = plt.subplot2grid((3,3),(1,0),colspan=2,rowspan=1)
ax3 = plt.subplot2grid((3,3),(1,2),colspan=1,rowspan=2)
ax4 = plt.subplot2grid((3,3),(2,0),colspan=1,rowspan=1)
ax5 = plt.subplot2grid((3,3),(2,1),colspan=1,rowspan=1)


plt.tight_layout()  
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# method 2:gridspec=> grid specification  网格规格
################################
plt.figure()   
#作第一张图
gs = gridspec.GridSpec(3,3)  
# : 在数字前代表到该数字结束    ：在数字后，代表至最后才结束
ax1 = plt.subplot(gs[0,:])   #第0行，所有列都占用了
ax2 = plt.subplot(gs[1,:2])  #第一行，占到2
ax3 = plt.subplot(gs[1:,2])  #水平 第一行开始，到最后一行    竖直：只是第二列
ax4 = plt.subplot(gs[2,0])
ax5 = plt.subplot(gs[2,1])


#作第二张图      从第1行0列的位置开始

plt.tight_layout()
plt.show()
```

# 13-图中图 plot in plot

```python
from matplotlib import pyplot as plt
import numpy as np

x = [1,2,3,4,5,6,7]
y = [1,3,4,2,5,8,6]

fig = plt.figure()
left,bottom,width,height=0.1,0.1,0.85,0.85

ax1 = fig.add_axes([left,bottom,width,height]) #通过直接增加坐标轴，而不是增加subplot子绘图
'''
plt.plot(x,y,'r')   #上下输出的内容一致
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
'''
ax1.plot(x,y,'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('title')

#内部坐标系1:  左上角的
left,bottom,width,height=0.2,0.6,0.25,0.25

ax2 = fig.add_axes([left,bottom,width,height])
ax2.plot(y,x,'b')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('title inside 1')

#内部坐标系2:  右下角的
left,bottom,width,height=0.65,0.2,0.25,0.25

ax2 = fig.add_axes([left,bottom,width,height])
ax2.plot(y[::-1],x,'g')   
print(y[::-1])   #[6, 8, 5, 2, 4, 3, 1]    将y列表的数据进行了反转
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('title inside 2')

plt.show()
```

# 14-次坐标 secondary axis

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(0,10,0.1)
#print(x)  #从0开始每隔0.1取一次
y1 = 0.05*x**2
y2 = -1*y1

fig,ax1=plt.subplots()
ax2 = ax1.twinx()  #将ax1轴关于x轴进行全反转，x轴不变，y轴全反转
ax1.plot(x,y1,'g-')  # -:代表直线
ax2.plot(x,y2,'b--') # --：代表虚线

ax1.set_xlabel('X data')
ax1.set_ylabel('Y2',color='g')
ax2.set_ylabel('Y1',color='b')


plt.show()
```
